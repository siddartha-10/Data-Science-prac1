## bmi below 18.5 is underweight 18.5 - 24.5 normal 25+ overweight

```python
catergories = ['underweight','normal','overweight']
def bmi(num):
    if(num < 18.5):
        return 'underweight'
    if(num >= 25):
        return 'overweight'
    if(num > 18.5 and num < 24.5):
        return 'normal'
df['bmi_categories'] = df['BMI'].apply(bmi)
```

```python
df.describe()
```

```python
df.info()
```

## Calculate average mental score (MentHlth) of subjects in each category of BMI (underweight, normal, overweight) (using for loop)!

```python
df.groupby('bmi_categories')['MentHlth'].mean()
```

## Find average physical health score (PhysHlth) of subjects in smokersâ€™ groups!

```python
def smoker(num):
    if(num==0):
        return 'non-smoker'
    else:
        return 'smoker'
df['smokers_categories'] = df['Smoker'].apply(smoker)
```

```python
df.head()
```

```python
df.info()
```

```python
df.groupby('smokers_categories')['PhysHlth'].mean()
```

## Compute max mental health score (MentHlth) in different categories of education level!

```python
def school(num):
    if(num==1):
        return 'kindergarten'
    if(num==2):
        return 'elementary'
    if(num==3):
        return 'high school'
    if(num==4):
        return 'high school grad'
    if(num==5):
        return 'college'
    if(num==6):
        return 'college graduate'
df['education_category'] = df['Education'].apply(school)
```

```python
df.info()
```

```python
df.head()
```

```python
df.groupby('education_category')['MentHlth'].max()
```

```python
df.groupby('education_category')['MentHlth'].mean()
```
